## 2. Data Preprocessing
The statistics from the above section shows that there is no necessary recaling or drop of any data entry of values. On 
the other hand, I do want to try doing clustering on transformed data. Inspired by SVM with inner product kernel, 
a nonlinear transformation to six dimensional data is applie to dataset 3, since it has so many wired shaped classes. 
The non linear transformation in to higher dimensional space might process a more separable case which should allevate 
the problem; however, the original version of dataset 3 will be still kept.

The transformed data has six dimensions, listed in below
x1 = 1
x2 = x^2
x3 = sqrt(2) * x * y
x4 = y^2
x5 = sqrt(2) * x
x6 = sqrt(2) * y

```python
x1 = (np.float_power(df_3['x'].values, 0)).reshape(-1,1)
x2 = (np.float_power(df_3['x'].values, 2)).reshape(-1,1)
x3 = (np.sqrt(2)*df_3['x'].values*df_3['y'].values).reshape(-1,1)
x4 = (np.float_power(df_3['y'].values,2)).reshape(-1,1)
x5 = (np.sqrt(2)*df_3['x'].values).reshape(-1,1)
x6 = (np.sqrt(2)*df_3['y'].values).reshape(-1,1)
data_transformed = np.concatenate([x1,x2,x3,x4,x5,x6],axis=1)
```

```python
data_transformed
print(data_transformed.shape)
df_3_transformed = pd.DataFrame(data=data_transformed, columns=['x1','x2','x3','x4','x5','x6'])
df_3_transformed['class'] = df_3['class']
```

## 3. Find Clusters
In this section we will use three clustering algorithms to cluster the three datasets
In this test we actively set the clusters as the number of classes of each dataset
### 3.1 df_3

```python
def silhouette_scatter(X,y,K):
    # silhouette plot and scatter plot
    fig = plt.figure(figsize=(12,4),dpi=100)
    ax1 = fig.add_subplot(1,2,1)
    ax2 = fig.add_subplot(1,2,2)
    for i in range(len(y)):
        if y[i] == -1:
            y[i] = 9999
    # the left plot is the silhouette plot
    ax1.set_xlim([-.1,1.0])
    vertical_spacing = (K+1)*10
    ax1.set_ylim([0, len(X)+vertical_spacing])
    silhouette_avg = silhouette_score(X, y)
    print('Silhouette score is {:.3f}'.format(silhouette_avg))
    sample_silhouette_values = silhouette_samples(X,y)
    y_lower = 10

    for i in range(K):
        ith_cluster_silhouette_values= sample_silhouette_values[y==i]
        ith_cluster_silhouette_values.sort()

        size_cluster_i = ith_cluster_silhouette_values.shape[0]
        y_upper = y_lower+size_cluster_i
        # color = plt.nipy_spectral(float(i) / K)
        color_map = cm.get_cmap('nipy_spectral')
        color = color_map(float(i)/K)
        ax1.fill_betweenx(np.arange(y_lower, y_upper),
                          0,
                          ith_cluster_silhouette_values,
                          facecolor=color,
                          edgecolor=color, 
                          alpha=0.7)
        ax1.text(-0.05, y_lower+0.5*size_cluster_i, str(i))

        y_lower = y_upper + 10
    ax1.set_title('Silhouette Plot')
    ax1.set_xlabel('Silhouette Score')
    ax1.set_ylabel("Cluster Label")
    ax1.axvline(x=silhouette_avg, color='red', linestyle='--')
    ax1.set_yticks([])
    ax1.set_xticks([-0.1,0,0.2,0.4,0.6,0.8,1])
    # the right one is the scatter plot
    color_map = cm.get_cmap('nipy_spectral')
    colors = color_map(y.astype(float)/K)
    ax2.scatter(X[:,0], X[:,1], marker='.',s=30, lw=0, alpha=0.7, c=colors, edgecolor='k')
    ax2.set_title("CLustered Data")
    ax2.set_xlabel("Feature space for the 1st feature")
    ax2.set_ylabel("Feature space fot the 2nd feature")

    plt.show()
```

```python
# kmeans clustering
K=6
X = df_3.drop(['class'],axis=1).values

km_3 = cluster.KMeans(n_clusters=K)
km_3_result = km_3.fit_predict(df_3.drop(['class'], axis=1))
y=km_3_result
silhouette_scatter(X,y,K)
df_3_result_km = df_3.copy()
df_3_result_km['predict'] = y
```

```python
K = 6
X_transformed = df_3_transformed.drop(['class'],axis=1).values
X = df_3.drop(['class'],axis=1).values
km_3_t = cluster.KMeans(n_clusters=K)
km_3_t_result = km_3.fit_predict(X_transformed)
y = km_3_t_result
silhouette_scatter(X,y,K)
df_3_result_km_t = df_3.copy()
df_3_result_km_t['predict'] = y
```

```python
# Agglomerative clustering
ac_3 = cluster.AgglomerativeClustering(n_clusters=6)
X = df_3.drop(['class'],axis=1).values
K=6
y = ac_3.fit_predict(X)
silhouette_scatter(X,y,K)
df_3_result_ac = df_3.copy()
df_3_result_ac['predict']=y
```

```python
# Agglomerative clustering
ac_3 = cluster.AgglomerativeClustering(n_clusters=6)
X = df_3.drop(['class'],axis=1).values
K=6
y = ac_3.fit_predict(df_3_transformed.drop(['class'], axis=1))
silhouette_scatter(X,y,K)
df_3_result_ac_t = df_3.copy()
df_3_result_ac_t['predict']=y
```

```python
# DBSCAN
X = df_3.drop(['class'],axis=1).values
db_3 = cluster.DBSCAN(eps=1.5).fit(X)
labels= db_3.labels_
print(labels)
K = len(set(labels))
if -1 in labels: K=K-1
silhouette_scatter(X, labels,K)
df_3_result_db = df_3.copy()
df_3_result_db['predict'] = labels
```

## 4 Evaluation

In the previous section we have already seen how those clustering algorithms are evaluated using silhouette plot, silhouette score and simlply eyeball the scatter plot.
In this section we use accuracy guided by the class label provided by the dataset

```python
def weighted_acc(df):
    possible_clusters = set(df['predict'])
    size_ = len(list(df.index))
    result = 0.0
    for cluster_id in possible_clusters:
        df_current = df.loc[df['predict'] == cluster_id]
        cluster_size = len(list(df_current.index))
        majority_class_size = 0
        classes = set(df['class'])
        for class_ in classes:
            df_current_class_specified = df_current.loc[df['class']==class_]
            num_obj = len(list(df_current_class_specified.index))
            if num_obj > majority_class_size:
                majority_class_size = num_obj
        result += float(majority_class_size)/float(size_)
    return result
```
